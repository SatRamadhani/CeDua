### Visualisasi urutan PTN berdasarkan anggaran BOPTN-nya.

```python
# Menghitung total BOPTN yang diterima PTN dari 2017-2021.
total_anggaran_boptn = anggaran_boptn.groupby(["nama"])["total_alokasi"].sum().rename("total").reset_index().sort_values("total", ascending=False)

# Plotting
fig, ax = plt.subplots(figsize=(30, 5))
total_anggaran_boptn.plot(ax=ax, x="nama", kind="bar")
```

#### Tahun 2019.

#### Tahun 2020.

```python
# Menghitung total BOPTN yang diterima PTN pada tahun 2020.
anggaran_boptn_2020 = anggaran_boptn[anggaran_boptn.tahun == 2020].groupby(["nama"])["total_alokasi"].sum().rename("total").reset_index().sort_values("total", ascending=False)

# PLotting
fig, ax = plt.subplots(figsize=(30, 5))
anggaran_boptn_2020.plot(ax=ax, x="nama", kind="bar")
```

### Visualisasi anggaran BOPTN berdasarkan rata-rata BOPTN 2017-2021.

```python
# Mencari peringkat berdasarkan rata-rata BOPTN dari tahun 2017-2021
peringkat_rerata_boptn = anggaran_boptn.groupby(["nama"])["total_alokasi"].mean().rename("avg").reset_index().sort_values("avg", ascending=False)
```

```python
# Buat atribut peringkat
peringkat_rerata_boptn = peringkat_rerata_boptn.assign(peringkat = [i for i in range (1,74)])
```

```python
peringkat_rerata_boptn
```

#### Visualisasi perubahan anggaran BOPTN Top 10 PTN dari tahun 2017 - 2021

```python
# Mencari top 10 BOPTN berdasarkan rata-rata BOPTN 2017-2021
top_10_boptn = peringkat_rerata_boptn.head(10)

# Plotting
top_10_plot = pd.merge(anggaran_boptn, top_10_boptn, how='inner',on=["nama"])
top_10_plot = top_10_plot.sort_values("avg", ascending=False)
top_10_plot = top_10_plot.drop("avg", axis=1)

fig, ax = plt.subplots(figsize=(15, 10))
sns.lineplot(ax=ax, x="tahun", y="total_alokasi", hue="nama", data=top_10_plot)
```

#### Visualisasi perubahan anggaran BOPTN Bottom 10 PTN dari tahun 2017 - 2021

```python
# Mencari bottom 10 BOPTN berdasarkan rata-rata BOPTN 2017-2021
bottom_10_boptn = peringkat_rerata_boptn.tail(10)

# Plotting
bottom_10_plot = pd.merge(anggaran_boptn, bottom_10_boptn, how='inner',on=["nama"])
bottom_10_plot = bottom_10_plot.sort_values("avg", ascending=False)
bottom_10_plot = bottom_10_plot.drop("avg", axis=1)

fig, ax = plt.subplots(figsize=(15, 10))
sns.lineplot(ax=ax, x="tahun", y="total_alokasi", hue="nama", data=bottom_10_plot)
```

### Selisih alokasi anggaran BOPTN dengan detail alokasi anggaran BOPTN

```python
# Karena pada dataset detail alokasi anggaran BOPTN hanya dari tahun 2019-2021, maka data tahun 2017 dan 2018
# di dataset alokasi anggaran BOPTN dihapus
alokasi_19_21 = anggaran_boptn.copy()
alokasi_19_21.drop(alokasi_19_21[(alokasi_19_21.tahun < 2019)].index, inplace = True)
alokasi_19_21
```

```python
# Mencari selisih
selisih_alokasi_boptn = pd.merge(alokasi_19_21, detail_boptn_tahun, how='inner',on=["nama", "tahun"])
selisih_alokasi_boptn['selisih'] = selisih_alokasi_boptn['total_alokasi'] - selisih_alokasi_boptn['total_detail']
selisih_alokasi_boptn
```

#### Mencari PTN yang memiliki selisih

```python
plot_selisih = selisih_alokasi_boptn[selisih_alokasi_boptn.selisih != 0].sort_values("selisih", ascending = False)
plot_selisih

# Ditemukan anomali selisih berniali minus
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.barplot(ax=ax, x="tahun", y="selisih", hue="nama", data=plot_selisih)
```
